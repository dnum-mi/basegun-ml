L'objectif de ce code est de créer un dictionnaire de calibre permettant de retrouver le calibre CIP a partir de n'importe quel synonyme. Pour cela nous nous sommes appuyés sur les connaissances des experts qui ont complétés un document de synonymes de calibres.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('CalibresListes2609.csv',on_bad_lines='skip')
#Données d'entrées, CSV complété par les experts des calibres CIP issus du RGA avec les synonymes connus de ces calibres
```

```python
data=data.drop(["referencesRGAAssoci�es"],axis=1)
data.replace(np.nan, '', inplace=True)
```

# Creation of matching table for Caliber
Obj -> pour chaque synonyme remonter au calibre CIP
PB:
Plusieurs calibres CIP pour même synonymes

Piste -> création dictionnaire avec clé : synonymes + CIP et value : liste calibres CIP associés si plusieurs


```python
def processRow(data,i,dictCalibre):
    for key in data:
        syn=data[key][i]
        if syn != '':
            try:
                dictCalibre[syn].append(data['calibreCanonUn'][i]) #si la valeur existe dans le dico
            except:
                dictCalibre[syn]=[data['calibreCanonUn'][i]]

    return dictCalibre
```

```python
from tqdm import tqdm
dictCalibre={}
for i in tqdm(range(len(data))):
    dictCalibre=processRow(data,i,dictCalibre)
```

```python
for key in list(dictCalibre.keys()):
    new_key=key.split('(')[0]
    dictCalibre[new_key] = dictCalibre[key]
    if new_key != key:
        del dictCalibre[key]
    else:
        pass
```
